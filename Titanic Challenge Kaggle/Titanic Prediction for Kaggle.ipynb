# Data Cleaning

```python
def nametitle(name):
    s = name.split(', ')
    t = s[1].split('.')
    return t[0]
```

```python
train['Name'] = train['Name'].apply(nametitle)
test['Name'] = test['Name'].apply(nametitle)
```

```python
print(train.head())
print(test.head())
```

# Handling Categorical Variables

```python
sex= pd.get_dummies(train['Sex'] ,drop_first=True)
sex1= pd.get_dummies(test['Sex'] ,drop_first=True)
```

```python
embark = pd.get_dummies(train['Embarked'] ,drop_first=True)
embark1 = pd.get_dummies(test['Embarked'] ,drop_first=True)
```

```python
train['Name'].value_counts()
test['Name'].value_counts()
```

```python
nam = pd.get_dummies(train['Name'] ,drop_first=True)
nam1 = pd.get_dummies(test['Name'] ,drop_first=True)
```

```python
nam.head()
```

```python
nam1.head()
```

```python
train = pd.concat([train,sex,embark,nam],axis=1)
test = pd.concat([test,sex1,embark1,nam1],axis=1)
```

```python
train.head()
test.head()
```

```python
train.drop(['Sex','Embarked','Name'],axis=1,inplace=True)
test.drop(['Sex','Embarked','Name'],axis=1,inplace=True)
```

```python
test.isnull().values.sum()
```

```python
me = np.mean(test['Fare'])
```

```python
#test['Embarked'].value_counts()
test['Fare'].fillna(me,inplace=True)
```

```python
sns.heatmap(train.isnull(),yticklabels=False)
```

```python
sns.heatmap(test.isnull(),yticklabels=False)
```
